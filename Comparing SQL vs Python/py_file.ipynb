<p> Corresponding SQL Code </p>



```python
df.groupby('occupation').size().to_frame('count').reset_index()

# to_frame: converts series to dataframe
```

<p> Corresponding SQL Code </p>



```python
df.groupby('weather')['temperature'].mean().to_frame('avg_mean').reset_index()
```

<p> Corresponding SQL Code </p>



```python
df.groupby('weather')['temperature'].nunique().to_frame('count_dist_temp').reset_index()
```

<p> Corresponding SQL Code </p>



```python
# min and max temperature
df.groupby('weather')['temperature'].min().to_frame('min_temp_by_weather').reset_index()
```

```python
df.groupby('weather')['temperature'].max().to_frame('max_temp_by_weather').reset_index()
```

<p> SQL Code </p>



```python
df.groupby("occupation").filter(lambda x: x["occupation"].iloc[0] == "Student").groupby('occupation').size()
```
